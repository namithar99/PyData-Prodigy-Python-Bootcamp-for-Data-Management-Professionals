```python
#Line Plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = pd.read_csv('C:/Users/hp/Downloads/iris-flower-dataset.csv')

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', hue='species', marker='o')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

```python
#Bar Plot

# Calculate the average sepal length for each species
avg_sepal_length = iris.groupby('species')['sepal_length'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_sepal_length, x='species', y='sepal_length', palette='viridis')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.show()
```

```python
#The average helps to condense this data into a single value that represents the central tendency for each category. 

import matplotlib.pyplot as plt

# Assuming avg_sepal_length is already calculated
plt.bar(avg_sepal_length['species'], avg_sepal_length['sepal_length'])

plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='sepal_length', hue='species', kde=True, palette='Set2', bins=15)
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the correlation matrix
# Select only numeric columns
numeric_columns = iris.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

print(corr_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

```python
import pandas as pd

# Load the employee dataset
employee_data = pd.read_csv('C:/Users/hp/Downloads/employee_data.csv')  # Adjust the file path if needed

print(employee_data)
```

```python
# Step 2: Fill missing values
# Example: Fill 'Age' with the median age and 'Department' with the most frequent value
employee_data['Age'] = employee_data['Age'].fillna(employee_data['Age'].median())
employee_data['Department'] = employee_data['Department'].fillna(employee_data['Department'].mode()[0])
```

```python
# Step 3: Filter data
# Example: Filter employees older than 30 years and earning a salary greater than 50,000
filtered_data = employee_data[(employee_data['Age'] > 30) & (employee_data['Salary'] > 50000)]
print("\nFiltered Data:")
print(filtered_data)
```

```python
# Step 4: Compute the average salary by department
avg_salary_by_department = employee_data.groupby('Department')['Salary'].mean()
print("\nAverage Salary by Department:")
print(avg_salary_by_department)
```

```python
# Total NaN values
df.isnull().sum()
```

```python
# Since we cannot know about all the directors,cast,country 
## So we fill invalid or something like that 
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df.isnull().sum()
```

```python
df_new=df.dropna()
df
```

```python
df_new.isnull().sum() #to check still null exist
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df_new, x='type', palette='viridis')
plt.title('Count of Movies vs TV Shows', fontsize=16)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
genres = df_new['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')
plt.title('Top 10 Genres', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
rating_counts = df_new['rating'].value_counts()
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='magma')
plt.title('Ratings Distribution', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
df = pd.read_csv('C:/Users/hp/Downloads\heart_disease.csv')
df
```

```python
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing numerical values with the median
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Encode categorical variables
df = pd.get_dummies(df, drop_first=True)

# Display the first few rows of the cleaned dataset
df.head()
```

```python
# Pair plot
sns.pairplot(df, hue='Heart Disease Status_Yes')
plt.show()
```

```python
# Distribution of Age
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```
